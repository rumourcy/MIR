#### 基于节拍的色度特征提取

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load('../data/music/Queen/Queen - Flash.mp3', duration=10)

hop_length = 512

y_harmonic, y_percussive = librosa.effects.hpss(y)

tempo, beat_frames = librosa.beat.beat_track(y=y_percussive, sr=sr)
print('Tempo:%fBPM'%(tempo))
beat_times = librosa.frames_to_time(beat_frames, sr=sr)
librosa.display.waveplot(y_harmonic, sr=sr)
plt.vlines(beat_times, -1, 1, color='r', alpha=0.5, linestyle='--', label='Beats')

chromagram = librosa.feature.chroma_cqt(y=y_harmonic, sr=sr)
plt.figure()
plt.subplot(2,1,1)
librosa.display.specshow(chromagram, y_axis='chroma', x_axis='time')
plt.title('chromagram')
plt.colorbar()
plt.tight_layout()

beat_chroma = librosa.util.sync(chromagram, beat_frames, aggregate=np.median)

plt.subplot(2,1,2)
librosa.display.specshow(beat_chroma[:, :-1], y_axis='chroma', x_axis='time',
                         x_coords=beat_times)
plt.vlines(beat_times, -1, 1, color='r', alpha=0.5, linestyle='--', label='Beats')
plt.title('beat_chroma')
plt.colorbar()
plt.tight_layout()
plt.show()
```

```python
def load_data(data_path):
    data = pd.read_csv(data_path, header=None).values
    features = data[:, 0:12].astype(np.float64)
    targets = data[:, 12]
    return features, targets

import time
from sklearn.model_selection import train_test_split

features, targets = load_data('chroma_cens.csv')
train_data, test_data, train_target, test_target = train_test_split(
    features, targets, test_size=0.2, random_state=int(time.time()))

from sklearn import svm

clf = svm.SVC()
clf.fit(train_data, train_target)
acc = clf.score(test_data, test_target)
print('test_acc:%f' % (acc))
```

```python
y, sr = librosa.load('../data/test/em.wav', duration=2)
chroma_cens = librosa.feature.chroma_cens(y=y, sr=sr)
chords = clf.predict(chroma_cens.T)
res = {}
for chord in chords:
    if res.has_key(chord):
        res[chord] += 1
    else:
        res[chord] = 1
num = float(len(chords))
for key in res:
    print('%s: %f' % (key, res[key]/num))
```
