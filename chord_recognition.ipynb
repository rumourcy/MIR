```python
from madmom.audio.chroma import DeepChromaProcessor
dcp = DeepChromaProcessor()
```

```python
from madmom.features.chords import DeepChromaChordRecognitionProcessor
decode = DeepChromaChordRecognitionProcessor()
```

```python
chroma = dcp('../data/music/The Beatles/Let It Be/Let It Be.mp3')
```

```python
segs = decode(chroma)
```

```python
import pandas as pd
pd.DataFrame(segs)
```

```python
times = []
chords = []
for seg in segs:
    times.append(seg[0])
    chords.append((seg[0],seg[2]))
```

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

duration = 10
y, sr = librosa.load('../data/music/The Beatles/Let It Be/Let It Be.mp3', duration=duration)

fig = plt.figure()
ax = fig.add_subplot(111)
librosa.display.waveplot(y, sr=sr)
plt.vlines(times, -1, 1, color='r', alpha=0.5, linestyle='--')
for chord in chords:
    if chord[0] < duration:
        ax.text(chord[0]+0.1, 0.8, chord[1], fontsize=15)
plt.show()
```
