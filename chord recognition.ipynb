```python
import os
import librosa
import pandas as pd

def extractFeature(filepath, type=None):
    pathDir =  os.listdir(filepath)
    for allDir in pathDir:
        child = os.path.join('%s/%s' % (filepath, allDir))
        if os.path.isfile(child):
            print(child)
            y, sr = librosa.load(child)
            chroma_cens = librosa.feature.chroma_cens(y=y, sr=sr)
            data_frame = pd.DataFrame(chroma_cens).T
            data_frame[12] = type
            data_frame.to_csv('chroma_cens.csv', mode='a+', header=False, index=False)
        else:
            extractFeature(child, allDir)
        
print("--------------Feature Extract----------------")
extractFeature('../data')
```

#### 音乐节拍检测

```python
import librosa
import librosa.display
import numpy as np
import matplotlib.pyplot as plt

y, sr = librosa.load(librosa.util.example_audio_file(), duration=10)
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
times = librosa.frames_to_time(beats, sr=sr)
print(times)

plt.figure()
librosa.display.waveplot(y, sr=sr)
plt.vlines(times, -1, 1, color='r', alpha=0.5, linestyle='--', label='Beats')
plt.show()
```

#### 基于节拍的色度特征提取

```python
import numpy as np
import librosa
import librosa.display
import matplotlib.pyplot as plt

y, sr = librosa.load(librosa.util.example_audio_file())

hop_length = 512

y_harmonic, y_percussive = librosa.effects.hpss(y)

tempo, beat_frames = librosa.beat.beat_track(y=y_percussive, sr=sr)
print('Tempo:%fBPM'%(tempo))
print('--------Beat_frames--------')
print(beat_frames)

chromagram = librosa.feature.chroma_cqt(y=y_harmonic, sr=sr)
plt.figure()
plt.subplot(2,1,1)
librosa.display.specshow(chromagram, y_axis='chroma', x_axis='time')
plt.title('chromagram')
plt.colorbar()
plt.tight_layout()

beat_chroma = librosa.util.sync(chromagram, beat_frames, aggregate=np.median)
beat_times = librosa.frames_to_time(beat_frames, sr=sr)

plt.subplot(2,1,2)
librosa.display.specshow(beat_chroma, y_axis='chroma', x_axis='frames')
plt.title('beat_chroma')
plt.colorbar()
plt.tight_layout()
plt.show()
```
